### Line plot

```python
# y in its range
x = np.arange(0, 1, 0.02)
y = x**2

plt.plot(y, label="basic plot - y in its range")
plt.legend()
```

```python
#y against x
plt.plot(x,y, label="basic plot - y against x")
plt.legend()
```

```python
# multiple in one
x = np.arange(0, 1, 0.02)
y1 = x**2
y2 = x**3

plt.plot(x,y1, label="basic plot - y1")
plt.plot(x,y2, label="basic plot - y2")
plt.legend()
```

```python
# some customization
plt.plot(x,y2, '.r', label="points highlighted");
plt.xlabel("input points")
plt.ylabel("output values")
plt.legend()
```

### SubPlot

```python
# subplot of 2x3
# The parameters are 
# - number of rows
# - number of columns
# - the index of the subplot (indices are counted row wise starting with 1) 
# The widths of the different subplots may be different. With use of GridSpec

plt.figure()
plt.subplot(2,3,1)
plt.plot(x,y2)
plt.xlabel("one")
plt.subplot(2,3,2)
plt.plot(x,y2)
plt.xlabel("two")
plt.subplot(2,3,3)
plt.plot(x,y2)
plt.xlabel("three")
plt.subplot(2,3,5)
plt.plot(x,y2)
plt.xlabel("five")
```

### Scatter Plot


```python
mean = [0, 0]
cov = [[2,4], [5, 9]]
xn, yn = np.random.multivariate_normal(mean, cov, 100).T
plt.scatter(xn,yn, label="xy distribution"); 
plt.legend()
```

### Histogram

```python
mean = [0, 0]
cov = [[2,4], [5, 9]]
xn, yn = np.random.multivariate_normal(mean, cov, 100).T

# histogram of the distribution on the x values
plt.hist(xn,bins=20, label="dist on x axis"); 
plt.ylim(0, 15)
plt.legend()
```

```python
mean = [0, 0]
cov = [[2,4], [5, 9]]
xn, yn = np.random.multivariate_normal(mean, cov, 100).T

# histogram on the distribution on the y values
plt.hist(yn,bins=30, label="dist on y axis"); 
plt.ylim(0, 15)
plt.legend()

# clearly there is more variance in the y-axis than in the x-axis
```

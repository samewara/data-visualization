```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
probs = np.full((6), 1/6)
face = [1,2,3,4,5,6]
plt.bar(face, probs)
plt.ylabel('Probability')
plt.xlabel('Fair Dice Roll')
plt.title('Uniform Distribution')
plt.grid()
axes = plt.gca()
axes.set_ylim([0,0.25])
```

```python
n = np.random.uniform(low=1, high=10,size=100)
plt.plot(n)
plt.title('Uniform Distribution - Normal data')
plt.grid()
```

```python
import numpy as np
mu, sigma = 0, 1
n = 100

s = np.random.normal(mu, sigma, n)
count, bins, ignored = plt.hist(s, 30, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
```

```python
for lambd in range(1,20, 4):
    x = np.arange(0, 10, 0.1)
    y = 0.2*lambd*np.exp(-0.1*lambd*x)

    plt.xlabel('Random Variable')
    plt.ylabel('Probability')
    plt.title("Exponential Distribution varying ")
    plt.legend()
    plt.grid()

    plt.plot(x,y, label=" = {:f}".format(0.1*lambd))
```
